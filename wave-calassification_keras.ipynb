```python
# from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation, SimpleRNN, Dense, LSTM
from keras.optimizers import Adadelta
from keras.utils import np_utils
from keras.preprocessing import sequence
import numpy as np
from IPython.display import SVG
from helper import DemoGenerator, myplot

from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
np.random.seed(13)
```

```python
dg = DemoGenerator()
x_lst, y_lst = dg.generate(size=100)
maxlen = 500
#print(x_lst[0])

train_X = sequence.pad_sequences(x_lst, maxlen=maxlen, truncating='post', dtype='float32')
train_Y = sequence.pad_sequences(y_lst, maxlen=maxlen, truncating='post')
#print(train_Y[0])
print(train_X.shape)

train_X = train_X[ :, :, np.newaxis]
train_Y = train_Y[ :, :, np.newaxis]

print(train_X.shape)
print(train_Y.shape)
```

```python

prv= 0
x = train_X[0][0:500] 
y = train_Y[0][0:500]
ctg =int(y[0])
print(ctg)
colors = ("tomato", "royalblue")
length= len(x)
print(length)
plt.plot(np.arange(0, 500), x, colors[1])
for idx in range(length):
    if int(y[idx]) != ctg:
        plt.plot(np.arange(prv, idx), x[prv:idx], colors[ctg])
        prv = idx

    
plt.show
```

```python

model = Sequential()
model.add(SimpleRNN(units=300, return_sequences=True, input_shape=(500, 1)))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='mean_squared_error',
              optimizer='adadelta',
              metrics=['accuracy'])

simple = model.fit(train_X, train_Y, batch_size=100, epochs=1000, validation_split=0.05)
```

```python
x = range(1000)
plt.plot(x, simple.history['acc'], label="simpleRNN")

plt.title("Accuracy (mean_squared_error)")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python

model = Sequential()
model.add(LSTM(units=300, return_sequences=True, input_shape=(500, 1)))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adadelta',
              metrics=['accuracy'])

simple_binary = model.fit(train_X, train_Y, batch_size=100, epochs=500, validation_split=0.05)
```

```python
x = range(500)
plt.plot(x, simple_binary.history['acc'], label="LSTM")

plt.title("binary train accuracy")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
